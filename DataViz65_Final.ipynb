```
# รวมคะแนนคะแนนสอบปากเปล่ากับคะแนนอาจารย์ที่ปรึกษา
mydata['คะแนนรวม'] = (mydata['คะแนนสอบปากเปล่า']+mydata['คะแนนอาจารย์ที่ปรึกษา'])
mydata
```

## ตัดเกรด

```
import numpy as np
```

```
mydata['Grade'] = np.where((mydata.คะแนนรวม < 50),
                  'F',np.where((mydata.คะแนนรวม >= 50)&(mydata.คะแนนรวม <55),
                  'D',np.where((mydata.คะแนนรวม >= 55)&(mydata.คะแนนรวม <60), 
                  'D+', np.where((mydata.คะแนนรวม >= 60)&(mydata.คะแนนรวม <65),
                  'C', np.where((mydata.คะแนนรวม >= 65)&(mydata.คะแนนรวม <70), 
                  'C+',np.where((mydata.คะแนนรวม >= 70)&(mydata.คะแนนรวม <75),
                  'B',np.where((mydata.คะแนนรวม >= 75)&(mydata.คะแนนรวม <80),
                  'B+',np.where((mydata.คะแนนรวม >= 80)&(mydata.คะแนนรวม <=100), 
                  'A', 'No Marks'))))))))
```
